# How has the pandemic affected the GDP per capita of the world's richest countries?

The COVID pandemic has been a major blow to the economies of all countries, but how exactly has it affected each of the world's richest countries?

![](images/figure1.jpeg)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import csv file
df = pd.read_csv('data/UNdata_Export_2.csv', sep=';')
```

```python
# Choose EU countries
richest = ['United States', 'China (mainland)', 'Japan', 'Germany', 
           'United Kingdom of Great Britain and Northern Ireland']
```

```python
# Plot the graph
colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']
fig, ax = plt.subplots(1, figsize=(20,12), facecolor='darkgrey')
ax.set_facecolor('darkgrey')
for i, v in enumerate(richest):
    temp = df[df['Country or Area'] == v]
    plt.plot(temp.Year, temp.Value, color=colors[i], lw=2.5)
            
    plt.text(temp.Year.values[0]+0.02, 
             temp.Value.values[0], 
             '{:,.2f}'.format(temp.Value.values[0]))
    
    plt.text(temp.Year.values[1]-0.02, 
             temp.Value.values[1], 
             '{:,.2f}'.format(temp.Value.values[1]),
             va='bottom', ha='right')
    
    plt.text(temp.Year.values[2]-0.02, 
             temp.Value.values[2], 
             '{:,.2f}'.format(temp.Value.values[2]),
             va='center_baseline', ha='right')
    
plt.xlim(2016.8,2021)
plt.xticks([2018,2019, 2020])
yticks = plt.yticks()[0]
plt.yticks([])

# Grid
ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)

# Remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.title('GDP Per Capta\n', loc='left', fontsize=20)
plt.legend(richest, loc='upper left', frameon=False)

plt.show()
```

```python
# Plot a variant
colors = ['#769465', '#69BDE0', '#E06D5E', '#2F7694', '#94E069']
fig, ax = plt.subplots(1, figsize=(20,12))

for i, v in enumerate(richest):
    # get a single country from the list
    temp = df[df['Country or Area'] == v]
    # plot the lines
    plt.plot(temp.Year, temp.Value, 
             color=colors[i], lw=2.5, 
             marker='o', markersize=5)
    # replace large labels
    if v == 'United Kingdom of Great Britain and Northern Ireland':
        v = 'UK'
        
    # End label
    plt.text(temp.Year.values[0]+0.02, temp.Value.values[0], v)
    
    # Start label
    plt.text(temp.Year.values[2]-0.02, temp.Value.values[2], v, ha='right')
    
# x limits, x ticks, and y label 
plt.xlim(2017.5,2021.5)
plt.xticks([2018, 2019, 2020])
plt.ylabel('USD')

# Get y ticks, replace 1,000 with k, and draw the ticks
yticks = plt.yticks()[0]
plt.yticks(yticks[1:-1], ['{}k'.format(i/1000) for i in yticks[1:-1]])

# Grid
ax.xaxis.grid(color='black', linestyle='solid', which='both', alpha=0.9)
ax.yaxis.grid(color='black', linestyle='dashed', which='both', alpha=0.1)

# Remove spines
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.title('GDP Per Capta\n', loc='left', fontsize=20)

plt.show()
```
